```python
# 0D tensor (scalar)
t0 = tf.constant(7)
# 1D tensor (vector)
t1 = tf.constant([1,2,3,4])
# 2D tensor (matrix)
t2 = tf.constant([[1,2],[3,4]])
# 3D tensor(cube/stack of matrix)
t3 = tf.constant([[[1],[2]],[[3],[4]]])
print(t0.shape,t1.shape,t2.shape,t3.shape)
```

```python
x =tf.constant([[1.0,2.0],[4.0,5.0]])
y = tf.constant([[5.0,6.0],[7.0,8.0]])
```

```python
#element wise operation

print("Add:\n",x+y)
print("subtract:\n",x-y)
print("Multiply:\n",x*y)
```

```python
#matrix multiplicaton 
print("Matrix Multiplication:\n",tf.matmul(x,y))
```
